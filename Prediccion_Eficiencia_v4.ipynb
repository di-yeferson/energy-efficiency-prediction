## 2.2 Funciones a utilizar
En esta parte se definen funciones auxiliares para procesar y analizar los datos. Incluyen herramientas para alinear texto, interpretar colinealidad (VIF) y clasificar coeficientes del modelo.

```python
# Funci√≥n para alinear datos de texto en el dataframe
def alinear_variable(df, columnas):
    """    Rellena con espacios a la derecha los valores de las columnas.
    Par√°metros:
    - df      : DataFrame de pandas
    - columnas: lista de nombres de columnas
    Retorna:
    - El mismo dataFrame con las columnas alineadas    """
    for col in columnas:
        if col in df.columns:
            # Calcula la longitud m√°xima de los valores en la columna
            max_len = df[col].astype(str).str.len().max()

            # Rellena con espacios a la derecha para igualar la longitud
            df[col] = df[col].astype(str).apply(lambda x: x.ljust(max_len))
    return df

# Funci√≥n para interpretar el VIF (detecta el nivel de colinealidad entre variables).
def interpretar_vif(valor):
    if   valor < 5   :  return '‚úÖ Sin colinealidad significativa'
    elif valor < 10  :  return '‚ö†Ô∏è Colinealidad moderada'
    elif valor < 30  :  return 'üü° Colinealidad Alta'
    elif valor < 1000:  return 'üî¥ Colinealidad Muy Alta'
    else             :  return 'üö® Colinealidad Extrema'


# Funci√≥n para clasificar la fuerza e impacto (positivo o negativo) de un coeficiente en el modelo.
def clasifica_coef(value):
    abs_val = abs(value)
    if   abs_val >= 5   :
      strength , emoji = 'fuerte  ' , 'üî¥' if value < 0 else 'üü£'
    elif abs_val >= 1   :
      strength , emoji = 'moderado' , 'üü†' if value < 0 else 'üîµ'
    elif abs_val >= 0.1 :
      strength , emoji = 'd√©bil   ' , 'üü°' if value < 0 else 'üü¢'
    else                :
      strength , emoji = 'm√≠nimo  ' , '‚ö™'

    direction = 'positivo' if value > 0 else 'negativo' if value < 0 else 'nulo'
    return emoji, f'{direction} {strength}'
```

##3.1. An√°lisis de colinealidad

El an√°lisis de colinealidad entre las variables se efect√∫a con la validacion VIF

  *  VIF = 1: Esto indica que no hay multicolinealidad
  *  VIF entre 1 y 5: Esto sugiere una multicolinealidad moderada
  *  VIF > 5: Existe una alta multicolinealidad
  *  VIF > 10: Esto indica una multicolinealidad grave.

### a. Determinar la colinealidad

```python
#***********************************************************************
#          Validacion VIF (Factor de Inflaci√≥n de la Varianza)
#***********************************************************************
# Se evaluar√° las variables de entrada
X = data1.copy()
X = X.drop(['Carga de Calefacci√≥n','Carga de Enfriamiento'], axis = 1)

# Agregar constante para el modelo
X_const = add_constant(X)

# Calcular VIF para cada variable
vif             = pd.DataFrame()
vif['Variable'] = X_const.columns
vif['VIF'     ] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]
vif_sin_const = round(vif[vif['Variable'] != 'const'],2)

# Agregar columna de interpretaci√≥n
vif_sin_const['Interpretaci√≥n'] = vif_sin_const['VIF'].apply(interpretar_vif)

# Mostrar tabla
df = alinear_variable(vif_sin_const, ["Variable","Interpretaci√≥n"])
print(vif_sin_const)
```

**Comentario**:

Las variables "Superficie", "√Årea de Pared" y "√Årea de Techo" tienen muy altamente colinealidad porque la validaci√≥n VIF resulta infinito.

## 4.7. Validaci√≥n Leave-One-Out (LOO)


*   Entrenar el modelo
*   Predecir el modelo
*   Calcular las m√©tricas:
    * Puntaje R2
      * 1 = perfecto
      * 0 = no explica nada
    * Ajuste R2
      * 1 = Ajuste excelente
      * 0.75 a 0.99 = Muy buen ajuste
      * 0.50 a 0.74 = Ajuste moderado
      * 0.25 a 0.49 = Ajuste d√©bil
      * menos de 0.25 = Ajuste muy pobre
    * MSE  Error cuadr√°tico medio (Desviaci√≥n Est√°ndar)
      * Cuanto m√°s bajo mejor
    * RMSE Ra√≠z del error cuadr√°tico medio (Error Est√°ndar)
    * MAE Error absoluto medio

```python
# Definir la funci√≥n que utiliza la validacion leave-one-out
# (dejar uno fuera) para evaluar los modelos.
def loo_results(model, X_scaled, y):
    # asignci√≥n de la funcion de validaci√≥n LeaveOneOut
    loo             = LeaveOneOut()
    test_array      = np.array([])
    pred_array      = np.array([])
    for train_index, test_index in loo.split(X_scaled):
        # Escalado o estandarizaci√≥n
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        y_train, y_test = y       [train_index], y       [test_index]

        # Entrenamiento
        model.fit(X_train, y_train)

        # Predicci√≥n
        y_pred          = model.predict(X_test)

        # Guardar resultados
        test_array      = np.append(test_array, y_test)
        pred_array      = np.append(pred_array, y_pred)

    # C√°lculo del coeficiente de determinacion R¬≤
    r2 = r2_score(test_array, pred_array)

    # C√°lculo de R¬≤ ajustado
    adjusted_r2 = (1 - (1-r2)*(len(test_array)-1)/(len(test_array)-X_scaled.shape[1]-1))

    # C√°lculo del error cuadr√°tico medio (MSE)
    mse  = mean_squared_error(test_array, pred_array)

    # C√°lculo de la raiz cuadrada del error cuadratico medio (RMSE)
    rmse = np.sqrt(mean_squared_error(test_array, pred_array))

    # C√°lculo del error absoluto medio (MAE)
    mae  = mean_absolute_error(test_array, pred_array)

    return adjusted_r2, mse, rmse, mae
```

## 4.8. Resultado de la validaci√≤n LOO

# <font color="yellow">5. Validaci√≥n de los modelos seleccionados</font>

El modelo es evaluado utilizando la m√©trica llamada Ajuste $R^2$ (R cuadrado)





$
R^2 ajustado= 1 - \left[ \left( 1 - R^2 \right) \cdot \frac{n - 1}{n - k - 1} \right]
$

Donde:
* $R^2$ = Coeficiente de determinaci√≥n
* n = tama√±o de la muestra
* k = n√∫mero de variables independientes

## 5.1. Dividir el dataset

```python
# Dividir la data en conjuntos de entrenamiento y prueba
X_train,X_test,y1_train,y1_test,y2_train,y2_test = train_test_split(
                                                      X, y1, y2,
                                                      test_size=0.2,
                                                      random_state=seed)
# Estandarizar la data
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

## 5.2. Funci√≥n para evaluar los modelos

```python
# Funci√≥n a evaluar para todos los modelos utilizando la data de test
def model_test(model, X_train, y_train, X_test, y_test):
    y_pred = model.fit(X_train, y_train).predict(X_test)
    # C√°lculo del coeficiente de determinacion R¬≤
    r2 = r2_score(y_test, y_pred)

    # M√©tricas de evaluaci√≥n:
    # C√°lculo de R¬≤ ajustado
    adjusted_r2=(1-(1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))

    # C√°lculo del error cuadr√°tico medio (MSE)
    mse  = mean_squared_error(y_test, y_pred)

    # C√°lculo de la raiz cuadrada del error cuadratico medio (RMSE)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    # C√°lculo del error absoluto medio (MAE)
    mae  = mean_absolute_error(y_test, y_pred)

    return adjusted_r2, mse, rmse, mae
```

## 5.3. Listar los modelos

```python
# Definir los modelos a evaluar
# - Regresi√≥n de √°rbol de decisiones
# - Regresi√≥n de √°rbol aleatorio
# - Regresi√≥n de potenciador de gradiente
# - Regresi√≥n de K vecinos m√°s cercanos
models_test = [
   DecisionTreeRegressor(random_state=seed)     ,
   RandomForestRegressor(random_state=seed)     ,
   GradientBoostingRegressor(random_state=seed) ,
   KNeighborsRegressor()]
```

## 5.4. Predecir con los modelos

```python
# Crear dataframes para almacenar las m√©tricas

# Para Calefacci√≥n
df_heating_test = pd.DataFrame(
    columns=['Model_Heating','Test_Adj_R2','Test_MSE','Test_RMSE','Test_MAE'])
for model in models_test:
    adjusted_r2,mse,rmse,mae = model_test(model, X_train, y1_train, X_test, y1_test)
    df_heating_test = pd.concat(
        [df_heating_test,
         pd.DataFrame([[model.__class__.__name__,adjusted_r2,mse,rmse,mae]],
         columns=['Model_Heating','Test_Adj_R2','Test_MSE','Test_RMSE','Test_MAE'])
        ], axis=0)
df_heating_test = df_heating_test.sort_values(by='Test_Adj_R2',
                     ascending=False).reset_index(drop=True)

# Para Enfriamiento
df_cooling_test = pd.DataFrame(columns=['Model_Cooling','Test_Adj_R2','Test_MSE',
                                        'Test_RMSE','Test_MAE'])
for model in models_test:
    adjusted_r2, mse, rmse, mae = model_test(model, X_train, y2_train, X_test, y2_test)
    df_cooling_test = pd.concat(
        [df_cooling_test,
         pd.DataFrame([[model.__class__.__name__,adjusted_r2,mse,rmse,mae]],
         columns=['Model_Cooling','Test_Adj_R2','Test_MSE','Test_RMSE','Test_MAE'])],
         axis=0)
df_cooling_test = df_cooling_test.sort_values(by='Test_Adj_R2',
                     ascending=False).reset_index(drop=True)

# Mostrar los resultados
print(" "*15, "M√©tricas de los modelos para Calefacci√≥n", "\n" , "‚îÄ"*70)
print(df_heating_test.round(4) , "\n")
print(" "*15, "M√©tricas de los modelos para Calefacci√≥n", "\n" , "‚îÄ"*70)
print(df_cooling_test.round(4))
```

# <font color="yellow">6. Presentaci√≥n de la etiqueta de eficiencia</font>

## 6.1 Funci√≥n para imprimir la etiqueta

```python
def etiqueta_eficiencia(letra, valor):
    letras  = ['A'      ,'B'      ,'C'      ,'D'      ,'E'      ,'F'      ,'G'      ]
    colores = ['#00B050','#92D050','#FFFF00','#FFC000','#FF9900','#FF0000','#A00000']
    largos  = [180      ,200      ,220      ,240      ,260      ,280      ,300      ]

    # Invertir para que A est√© arriba
    letras  = letras [::-1]
    colores = colores[::-1]
    largos  = largos [::-1]

    fig, ax = plt.subplots(figsize=(6, 4))
    for i, (l, c, w) in enumerate(zip(letras, colores, largos)):
        y     = i
        h     = 0.8
        punta = 20  # largo de la punta triangular

        # Coordenadas del pol√≠gono tipo flecha
        coords = [
            [0        , y - h/2],
            [w - punta, y - h/2],
            [w        , y      ],
            [w - punta, y + h/2],
            [0        , y + h/2]
        ]
        flecha = Polygon(coords, closed=True, facecolor=c, edgecolor='black')
        ax.add_patch(flecha)
        ax.text(5, y, f'{l}', va='center', ha='left', fontsize=12,
                fontweight='bold', color='black')

    # Marca el valor actual
    idx = letras.index(letra.upper())
    ax.plot(valor, idx, 'ko', markersize=10)
    ax.text(valor + 10, idx, f'{valor:.1f} kWh/m¬≤', va='center', fontsize=10)

    ax.set_xlim(0, max(largos) + 50)
    ax.set_ylim(-0.5, len(letras) - 0.5)
    ax.set_yticks([])
    ax.set_title('Etiqueta de Eficiencia Energ√©tica \n', fontsize=14)
    plt.tight_layout()
    plt.show()
```

## 6.2 Funci√≥n para predecir la eficiencia energ√©tica

```python
def predice_eficiencia(dato,scaler,dt_heating,rf_cooling):
    dato_escalado = scaler.transform(dato)
    pred_calefaccion  = dt_heating.predict(dato_escalado)[0]
    pred_enfriamiento = rf_cooling.predict(dato_escalado)[0]
    indice = pred_calefaccion + pred_enfriamiento
    if   indice < 50   :  indice_txt= 'A' #'A-Muy eficiente'
    elif indice < 75   :  indice_txt= 'B' #'B-Eficiente'
    elif indice < 100  :  indice_txt= 'C' #'C-Moderadamente eficiente'
    elif indice < 150  :  indice_txt= 'D' #'D-Media'
    elif indice < 225  :  indice_txt= 'E' #'E-Poco eficiente'
    elif indice < 300  :  indice_txt= 'F' #'F-Ineficiente'
    else               :  indice_txt= 'G' #'G-Muy ineficiente'
    print('Carga de Calefacci√≥n  = ',round(pred_calefaccion ,1), 'kWh/m¬≤')
    print('Carga de Enfriamiento = ',round(pred_enfriamiento,1), 'kWh/m¬≤')
    print('√çndice                = ',round(indice,1)           , 'kWh/m¬≤')
    print('Letra                 = ',indice_txt)
    print('')
    etiqueta_eficiencia(letra=indice_txt, valor=indice)
```
